```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
tq = TweetTokenizer()
```

```python
data_stem=dataD
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
stemmer = PorterStemmer()
nuevoD = []
for linea in data_stem:
    t=tq.tokenize(linea)
    auxL=""
    for pal in t:
        palabra=stemmer.stem(pal)
        auxL=auxL+palabra+" "
    linea=auxL
    nuevoD.append(linea)
    #print(linea)
print(nuevoD[1])    
```

```python
dataset["Description"] = nuevoD
```

```python
dataX=np.asarray(dataset["Description"])
data_stem = dataX
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(data_stem)
X.shape
```

```python
print(X)
```

```python
posicion = set(dataP)
print(posicion)
```

```python
pos = dataset["Position Level"]
num_posicion = []
for i in pos:
    if "Ejecutivo" == i:
        num_posicion.append(1);
    if "Gerencia Media" == i:
        num_posicion.append(2);
    if "Dirección / Alta Gerencia" == i:
        num_posicion.append(3);
    if "Asistente" == i:
        num_posicion.append(4);
    if "Docente" == i:
        num_posicion.append(5);
    if "Supervisor" == i:
        num_posicion.append(6);
    if "Analista" == i:
        num_posicion.append(8);
    if "Practicante" == i:
        num_posicion.append(9);
    if "Consultor / Asesor" == i:
        num_posicion.append(10);
    if "Operativo" == i:
        num_posicion.append(11);
    if "Profesional especializado" == i:
        num_posicion.append(12);
    if "Otros" == i:
        num_posicion.append(13);
#print (num_sent)
dataset["Position Level"] = num_posicion
dataset
```

```python
y = np.asarray(dataset["Position Level"])
y.shape
print(y)
```

```python
from sklearn.model_selection import train_test_split

(X_train,
 X_test,
 y_train,
 y_test) = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
from sklearn import metrics
from sklearn.linear_model import SGDClassifier
from sklearn import tree
from sklearn.neural_network import MLPClassifier

target_names = dataset['Position Level'].astype(str).unique()

classifiers = [
  #(KNeighborsClassifier(n_neighbors=4), 'kNN'),
   #(BernoulliNB(alpha=.01), "Bernoulli Naive Bayes"),
    (SGDClassifier(loss="hinge", penalty="l2"),"SGDClassifier"),
    #(tree.DecisionTreeClassifier(),"tree.DecisionTreeClassifier"),
    #(MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1),"MLPClassifier")
]

for clf, name in classifiers:
    print('Clasificador: {}'.format(name))
    clf.fit(X_train, y_train)
    pred = clf.predict(X_test)
    score = metrics.accuracy_score(y_test, pred)
    print('Score: {}'.format(score))
    print("Classification report:")
    print(metrics.classification_report(y_test, pred, target_names=target_names))
```
